```python
import matplotlib.pyplot as plt

# Reading the data
with open('Data_1.txt', 'r') as file:
    data_x = [float(line.strip()) for line in file.readlines()]

with open('Data_2.txt', 'r') as file:
    data_y = [float(line.strip()) for line in file.readlines()]

with open('Data_1.txt', 'r') as file:
    data_1 = [float(line.strip()) for line in file.readlines()]

with open('Data_2.txt', 'r') as file:
    data_2 = [float(line.strip()) for line in file.readlines()]

# Function to find maxima and minima
def find_peaks(data):
    maxima = []
    minima = []
    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            maxima.append(i)
        if data[i] < data[i - 1] and data[i] < data[i + 1]:
            minima.append(i)
    return maxima, minima

# Finding peaks for data_1 and data_2
maxima_1, minima_1 = find_peaks(data_1)
maxima_2, minima_2 = find_peaks(data_2)

# Writing maxima and minima to files
def write_peaks_to_file(maxima, minima, data, file_prefix):
    with open(f'{file_prefix}_maxima.txt', 'w') as max_file:
        for i in maxima:
            max_file.write(f'Index: {i}, Value: {data[i]}\n')

    with open(f'{file_prefix}_minima.txt', 'w') as min_file:
        for i in minima:
            min_file.write(f'Index: {i}, Value: {data[i]}\n')

write_peaks_to_file(maxima_1, minima_1, data_1, 'Data_1')
write_peaks_to_file(maxima_2, minima_2, data_2, 'Data_2')

# Plotting Data_1
plt.figure(figsize=(10, 4))
plt.plot(data_1, label='Data 1')
plt.scatter(maxima_1, [data_1[i] for i in maxima_1], color='red', label='Maxima')
plt.scatter(minima_1, [data_1[i] for i in minima_1], color='blue', label='Minima')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Data_1')
plt.legend()
plt.grid(False)
plt.show()

# Plotting Data_2
plt.figure(figsize=(10, 4))
plt.plot(data_2, label='Data 2')
plt.scatter(maxima_2, [data_2[i] for i in maxima_2], color='red', label='Maxima')
plt.scatter(minima_2, [data_2[i] for i in minima_2], color='blue', label='Minima')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Data_2')
plt.legend()
plt.grid(False)
plt.show()
```
